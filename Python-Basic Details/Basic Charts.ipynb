# seaborn

```python
import seaborn as sns
```

```python
# example of scatter plot
sns.scatterplot(x=x,y=y)
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.show()
```

```python
sns.scatterplot(x=x,y=y)
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.xlabel('X-label')
plt.ylabel('y-label')
plt.title("scatter plot")
plt.show()
```

```python
#line chart
x=[1,2,3,4,5]
y=[10,4,15,7,12]

plt.plot(x,y,marker='o',linestyle='-',color='g')

# marker='o' add circular marks at each data point.
# linestyle='-' specific a solid line.
# color set line color.

plt.show()
```

```python

x=[1,2,3,4,5]
y=[10,4,15,7,12]

plt.plot(x,y,marker='o',linestyle='-',color='g')

#adding labels and title
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.title('Line Chart Example')
plt.show()
```

```python
# bar chart
category=["category A","category B","category C","category D"]
values=[25,16,30,22]

plt.bar(category,values,color='g')
plt.show
```

```python
category=["category A","category B","category C","category D"]
values=[25,16,30,22]

plt.bar(category,values,color='g')

plt.xlabel('x-label')
plt.ylabel('y-label')
plt.title('Bar chart example')
plt.show()
```

```python
# histogram 

data=[1,2,2,4,4,2,6,3,5,7,1,3,8,9,4,6,3,5,6,4]

plt.hist(data,color='skyblue',edgecolor='black')
plt.show()
```

```python
data=[1,2,2,4,4,2,6,3,5,7,1,3,8,9,4,6,3,5,6,4]

plt.hist(data,bins=5,color='skyblue',edgecolor='black')

# bins:this parameter specific the number of bins (intervals) you want the data to be divided into more bins give amore detailed view of distribut
plt.show()
```

```python
data=[1,2,2,4,4,2,6,3,5,7,1,3,8,9,4,6,3,5,6,4]

plt.hist(data,bins=6,color='skyblue',edgecolor='black')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.title('histogram chart example')

plt.show()
```

```python
# pie chart

size=[30,28,15,10,20]
labels=["category A","category B","category C","category D","category E"]
colors=["gold","skyblue","lightcoral","lightgreen","orange"]

plt.figure(figsize=(4,4))
plt.pie(size,labels=labels,autopct='%1.1f%%',startangle=140,colors=colors)

# autopct='%1.1f%%' add percentage labels with one decimal place.
# startangle=140 sets the starting angle forthe first wedge.


# adding title 
plt.title('Pie Chart Example')
plt.show()
```

```python
labels=["category A","category B","category C","category D"]
sizes=[15,30,45,10]
plt.figure(figsize=(4,4))
plt.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=140,colors=['gold','skyblue','lightgreen','orange'])


plt.title('Pie Chart Example')
plt.show()
```

```python
#boxplot

g1=[10,15,22,26,30]
g2=[6,10,14,20,28]
g3=[15,18,21,28,35]

plt.boxplot([g1,g2,g3],label=['g1','g2','g3'])

plt.show()
```

```python
g1=[10,15,22,26,30]
g2=[6,10,14,20,28]
g3=[15,18,21,28,35]

plt.boxplot([g1,g2,g3],label=['g1','g2','g3'])
plt.title('Box plot Example')
plt.xlabel('Groups')
plt.ylabel('Values')


plt.show()
```

```python
#heatmap

data=[[1,2,3],
      [4,5,6],
      [7,8,9]]

sns.heatmap(data,annot=True)
# annot: a boolean or rectangular dataset.If True ,write the data value in each cell.
plt.show()
```
