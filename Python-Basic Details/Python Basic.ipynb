## Variable Naming

```python
# 1. Variable name must strat with a letter
Name="My name is Abc"
print(Name)

# 2. Variable name must strat with a underscores
_name="Abc"
print(_name)

# 3. Variable name can only contain alpha-numeric characters and underscores(A-z,0-9,and _)
name_1="xyz"
print(name_1)

nameab_10k="xyzabc"
print(nameab_10k)

# 4. Variable name are case-sensitive
Abc_Name="XYZ"
print(Abc_Name)
```

# Loops

## if-else statement
An if-else statement is used to make a decision. It checks a condition and executes one block of code if the condition is true, and another block of code if the condition is false.

```python
age=17
if age >=18:
    print("You are an adult.")
else:
    print("you are not an adult")
```

## nested loop

```python
scroe=78

if scroe>=90:
    grade='A'
elif scroe>=80:
    grade='B'
elif scroe>=70:
    grade="C"
else:
    grade="D"
    
print("Your grade is:",grade)
```

## while loop

A while loop repeats a block of code an unknown number of times until a condition is no longer met.

```python
# program to display numbers from 1 to 5
count=0
while count< 5:
    print("count is:",count)
    count +=1  # count= count+1
```

```python
# program to display numbers from 1 to 5

# intialize the variable
i=1
n=5

# while loop from i=1 to 5
while i<=n:
    print(i)
    i+=1
```

## for loop
A for loop is used to repeat a block of code for each item in a sequence, such as:
list
tuple
string
range of numbers

It helps avoid writing the same code again and again.

```python
abc=["apple","banana","kiwi"]
for ab in abc:
    print(ab)
```

```python
n=5
for i in range(0,n):
    print(i)
```
